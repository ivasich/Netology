```python
#p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

def p(document):
    for x in documents:
        if x['number'] == document:
            return x['name']
    assert False,"Отсутствует в списке"

document = input("Введите номер документа: ")
print(p(document))
```

```python
#s-shelf команда, которая спросит номер документа и выведет номер полки, на которой он находится

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def s(pass_num):
    for x,y in directories.items():
        if pass_num in y:
            return x
        assert False,"Отсутствует в списке"
pass_num = input("Введите номер документа: ")
print(f"Номер полки вашего документа: {s(pass_num)}.")


```

```python
#d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. 
#Предусмотрите сценарий, когда пользователь вводит несуществующий документ;

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def delete_shelf(num_doc):
    num_doc = input('Введите номер документа:')
    del_check = False
    for el in documents:
        if el['number'] == num_doc:
            documents.remove(el)
            del_check = True
    if del_check == False:
        print("Документа не существует")
    for key, value in directories.items():
        if num_doc in value:
            value.remove(num_doc)



delete_shelf(documents)

print(documents)
```

```python
#m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки 
#на целевую. 
#Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ 
#или переместить документ на несуществующую полку;

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]


directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def m(num_doc, shelf_num):
        doc_existence = False
    
    
        if shelf_num not in directories:
            return ('Полки не существует')
    
            for key, value in directories.items(): 
                if num_doc in value: 
                    doc_existence = True
                directories[shelf_num] += [num_doc]
        
        if doc_existence:
            return "Документ успешно перемещен"
        else :
            return "Документ не существует"

        
num_doc = input('Введите номер:')
shelf_num = input('Введите полку:')

print(m(num_doc, shelf_num))
print(directories)
```

```python
#a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. 
#Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку.

def add_info (document_number, d_type, name, shelf_number):
    '''
    Добавляет новый документ в каталог и в перечень полок.
    '''
    if shelf_number in directories.keys():
        a = directories[shelf_number]
        a.append(document_number)
        d1 = dict(d_type = d_type, number = document_number, name = name)
        documents.append(d1)
    else:
        print("Несуществующая полка")
        return

add_info ('111111', 'card', 'ilya', '3')

print(documents)
print(directories)
```
